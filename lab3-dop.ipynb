```python
def func(x):
    return (np.sin(x/10)+np.cos(x/100))

def func1(func, a, b):
    x = np.arange(a, b)
    y = func(x) 
    plt.plot(x, y)
    
    y_diff = np.diff(y)
    plt.plot(x[:-1], y_diff)
    
    res_max = []
    res_zero = []
    max_y = np.max(np.abs(y_diff))
    for i in range(len(y_diff)):
        if (y_diff[i] == max_y): 
            res_max.append(i)
        if (np.abs(y_diff[i]) < 0.00001): 
            res_zero.append(i)  
    
    plt.plot(x, y)
    for idx in res_max:
        plt.scatter(x[idx], y[idx], color='green')
    for idx in res_zero:
        plt.scatter(x[idx], y[idx], color='red')

func1(func, -1000, 1000)
```

```python
import pylab
from mpl_toolkits.mplot3d import Axes3D

def func(x, y):
    return np.sin(x) * np.sin(y)

def extr_func(z):
    x = np.arange(-3.14, 3.14, 0.1)
    y = np.arange(-3.14, 3.14, 0.1)
    z = func(x, y)

    local_mins = []  
    local_maxs = []  
    global_mins = []  
    global_maxs = []  

    min_z = np.min(z)
    max_z = np.max(z)

    for i in range(len(z)):
        if (z[i] == min_z):
            global_mins.append(i)

        if (z[i] == max_z): 
            global_maxs.append(i) 

        if (i > 0 and i < len(z)-1):
            if (z[i-1] > z[i] and z[i+1] > z[i]): 
                local_mins.append(i)  

        if (z[i-1] < z[i] and z[i+1] < z[i]):
            local_maxs.append(i) 

    xgrid, ygrid = np.meshgrid(x, y)
    zgrid = func(xgrid, ygrid)
    
    fig = pylab.figure()
    axes = Axes3D(fig)
    axes.plot_surface(xgrid, ygrid, zgrid)
    pylab.show()
    return local_mins, local_maxs, global_mins, global_maxs

print(extr_func(func))
```
